```python
def missing_values(dt):
    for i in dt.columns:
        if dt[i].dtype == 'object':
            print()
            print("Categorical column Characterstics:")
            print(("Column Name"),i)
            print("Missing values   " ,dt[i].isnull().sum())
            print(dt[i].value_counts())
            print("Mode",dt[i].mode())
            print("......................................")
        elif dt[i].dtype=='int32' or 'float64' or 'int62':
            print()
            print("Numerical column")
            print(("Column Name"),i)
            print("Missing values   " ,dt[i].isnull().sum())
            print(dt[i].describe())
            print(dt[i].value_counts())
            print("Mode",dt[i].mode())
            print("......................................")
```

```python
missing_values(data)
```

```python
def Uni_Analysis(dt):
    for i in dt.columns:
        if dt[i].dtype == 'object':
            print()
            print("Categorical column Characterstics:")
            print(("Column Name"),i)
            print(dt[i].value_counts())
            print("Mode",dt[i].mode())
            print(".....................................")
        elif dt[i].dtype=='float64':
            print()
            print("Numerical column")
            print(("Column Name"),i)
            print("Describing the data " , dt[i].describe())
            print("Mode",dt[i].mode())
            print("......................................")
        elif dt[i].dtype=='int32' or 'int62':
            print()
            print("Numerical column")
            print(("Column Name"),i)
            print("Min ",dt[i].min())
            print("Max ",dt[i].max())
            print(dt[i].value_counts())
            print("Mode",dt[i].mode())
            print("......................................")
```
